```python
# https://docs.python.org/2/library/re.html#re.compile
import re
# шаблон выборки 
# http://www.petefreitag.com/cheatsheets/regex/
pattern ='[^абвгдёжзийнпуфхцчшщъыьэюяАБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ-]' # фильтр букв не содержащихся в слове "ростелеком"
prog = re.compile( pattern )
```

```python
with open('dic_orf.txt', 'r') as f:  #открываем файл с словарем
    l1="111" # задаем произвольную строку
    i=1    
    for line in f:
        k3= line.find('#')  #определение конца слова в строке
        line = line[:k3]    # обрезка строки по первое слово
        if len(prog.findall( line ))==k3>2 and l1!= line:   # проверка условия совпадения с шаблоном, а также длины слова - более 2-х букв и повторения слов
            if prog.findall( line ).count('р') <2:      # проверка на количество букв  в строке
                if prog.findall( line ).count('о') <3:
                    if prog.findall( line ).count('с') <2:
                        if prog.findall( line ).count('т') <2:
                            if prog.findall( line ).count('е') <3:
                                if prog.findall( line ).count('л') <2:
                                    if prog.findall( line ).count('к') <2:
                                        if prog.findall( line ).count('м') <2:
                                            print (i, line )
                                            l1= line
                                            i+=1
                                                   
          
```
